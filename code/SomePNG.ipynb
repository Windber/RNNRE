# 解释sigmoid（x，a）函数的好处(可能存在的问题是梯度爆炸，但是尝试证明梯度有一个最大值）

```python
import matplotlib.pyplot as plt
def sigmoid_a(x, a):
    return 1 / ( 1 + torch.exp(-a * x))
x = torch.Tensor(range(-1000, 1000)) * 1e-1
plt.hold(True)
plt.plot(x.numpy(), sigmoid_a(x, 1).numpy())
plt.plot(x.numpy(), sigmoid_a(x, 10).numpy())
plt.plot(x.numpy(), sigmoid_a(x, 100).numpy())
plt.plot(x.numpy(), sigmoid_a(x, 1000).numpy())
```

```python
#x = torch.Tensor(range(-1000, 1000)) * 1e-2
def grad_a(x, a):
    return a* sigmoid_a(x, a) * (1 - sigmoid_a(x, a))
```

```python
plt.hold(True)
plt.plot(x.numpy(), grad_a(x, 1).numpy())
plt.plot(x.numpy(), grad_a(x, 10).numpy())
plt.plot(x.numpy(), grad_a(x, 100).numpy())
plt.plot(x.numpy(), grad_a(x, 1000).numpy())
```

# tanh梯度取值（0,1）

```python
x = torch.Tensor(range(-1000, 1000)) * 1e-2

tanh = torch.nn.Tanh()
y = tanh(x)


def grad_tanh(x):
    return 1 - torch.pow(tanh(x), 2)


ygx = grad_tanh(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```

```python
plt.plot(x.numpy(), ygx.numpy())
```
